```python
############################################################################
#                                                                          #
#            IMPLEMENTATION EN PYTHON DE QUELQUES TESTS NON PARAMETRIQUES  #
############################################################################

# TESTS D'ADEQUATION 
#Test d'adequation de kolmogorov-smirnov à la loi normale
# Importation du package et de la fonction de test
from scipy.stats import kstest
# Creation de la série de donnée
data = [0.873, 2.817, 0.121, -0.945, -0.055, -1.436, 0.360, -1.478, -1.637, -1.869]
#Application du test et recuperation des valeurs de retour
stat,p_valeur=  kstest(data,'norm')
# Affichage de la statistique et la p.value
print('statistique du test = %.3f    p.value= %.3f' % (stat,p_valeur))
# Fixation du seuil du test 
seuil=0.05
#Decision du test  
if p_valeur> seuil:
	print('On accepte H0 autrement les données suivent une loi normale')
else:
	print('On rejette H0')
```

```python
#Test d'adequation
#Test de shapiro-Wilk
# Importation du package et de la fonction de test
from scipy.stats import shapiro
# Creation de la série de donnée
data = [0.873, 2.817, 0.121, -0.945, -0.055, -1.436, 0.360, -1.478, -1.637, -1.869]
#Application du test et recuperation des valeurs de retour
stat, p = shapiro(data)
# Affichage de la statistique et la p.value
print('stat=%.3f, p=%.3f' % (stat, p))
# Fixation du seuil du test
seuil=0.05
#Decision du test 
if p >seuil:
	print('On accepte H0 autrement les données suivent une loi normale')
else:
	print('On rejette H0')
```

```python
##TESTS D'ECHANTILLONS INDEPENDANTS
## Independance du chi 2
# Importation du package et de la fonction de test
from scipy.stats import chi2_contingency
# Creation de la série de donnée
table = [[10, 20, 30],[6,  9,  17]]
#Application du test et recuperation des valeurs de retour
stat, p, dof, expected = chi2_contingency(table)
# Affichage de la statistique et la p.value
print('statistique du test=%.3f, p.value=%.3f' % (stat, p))
# Fixation du seuil du test
seuil=0.05
#Decision du test 
if p > seuil:
	print('On accepte H0 autrement les echantillons sont independants')
else:
	print('On rejette H0')
```

```python
##TESTS D'ECHANTILLONS INDEPENDANTS
#Mann-Whitney U test
# comparaison d'echantillons
# Importation du package et de la fonction de test
from scipy.stats import mannwhitneyu
# Creation des séries de données
data1=[10.3,10.8,12.3,13.8,13.8,14.3,15.3,15.8,16,17]
data2=[8.8,9.3,9.8,10.8,11.3,11.8,12.8,13.3,14.8,17]
#Application du test et recuperation des valeurs de retour
stat, p = mannwhitneyu(data1, data2)
# Affichage de la statistique et la p.value
print('Statistique du test =%.3f, p.value=%.3f' % (stat, p))
# Fixation du seuil du test
seuil = 0.05
#Decision du test 
if p > seuil:
	print('On accepte H0 autrement les echantillons sont independants')
else:
	print('On rejette H0')
```

```python
##TESTS D'ECHANTILLONS INDEPENDANTS
#test d'echantillons independants de Kolmogorov Smirnov
# Importation du package et de la fonction de test
from scipy.stats import ks_2samp
# Creation de la série de donnée
serie1=[19,30,20,19,29,25,21,24,50]
serie2=[25,21,17,15,14,14,22,17]
#Application du test et recuperation des valeurs de retour
stat,p=ks_2samp(serie1, serie2)
# Affichage de la statistique et la p.value
print('statistique du test=%.3f, p.value=%.3f' % (stat, p))
# Fixation du seuil du test
seuil=0.05
#Decision du test 
if p > seuil:
    print('On accepte H0 autrement les echantillons sont independants')
else:
	print('On rejette H0')
```

```python
##TESTS D'ECHANTILLONS INDEPENDANTS
#Kruskal-Wallis H-test
# comparaison de plusieurs echantillons
# Importation du package et de la fonction de test
from numpy.random import randn
from scipy.stats import kruskal 
#generation de donnnées aleatoires
data1 = randn(95)  
data2 = randn(100) 
data3 = randn(105)  
#Application du test et recuperation des valeurs de retour
stat, p = kruskal(data1, data2, data3)
# Affichage de la statistique et la p.value
print('Statistique du test=%.3f, p.value=%.3f' % (stat, p))
# Fixation du seuil du test
seuil = 0.05
#Decision du test 
if p > seuil:
	print('On accepte H0 autrement les echantillons sont independants')
else:
	print('On rejette H0')
```

```python
##TESTS D'ECHANTILLONS APPARIÉS
# Test du rang signé de Wilcoxon
# comparason d'echantillons
from scipy.stats import wilcoxon
# Creation de la série de donnée
data1=[6,5,3,4,5,4,9,6,5]
data2=[7,4,4,2,2,4,5,3,6]
#Application du test et recuperation des valeurs de retour
stat, p = wilcoxon(data1, data2)
# Affichage de la statistique et la p.value
print('Statistique du test=%.3f, p.value=%.3f' % (stat, p))
# Fixation du seuil du test
seuil = 0.05
#Decision du test 
if p > seuil:
	print('On accepte H0 autrement les echantillons sont identiques')
else:
	print('On rejette H0')
```

```python
##TESTS D'ECHANTILLONS APPARIÉS
# Test Friedman 
# comparason d'echantillons 
# Importation du package et de la fonction de test
from numpy.random import randn
from scipy.stats import friedmanchisquare
# Creation de la série de donnée
data1 = randn(95)  
data2 = randn(95) 
data3 = randn(95)
#Application du test et recuperation des valeurs de retour
stat, p = friedmanchisquare(data1, data2, data3)
# Affichage de la statistique et la p.value
print('Statistique du test=%.3f, p.value=%.3f' % (stat, p))
# Fixation du seuil du test
seuil = 0.05
if p > seuil:
	print('On accepte H0 autrement les echantillons sont identiques')
else:
	print('On rejette H0')
```
